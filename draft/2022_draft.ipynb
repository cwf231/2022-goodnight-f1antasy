```python
import pandas as pd
```

```python
drivers = pd.read_csv('../data/drivers.csv', index_col=0)
drivers = pd.merge(
    pd.read_csv('../data/results.csv', index_col=0)[['DriverID','ConstructorID']], 
    drivers,
    on='DriverID'
    )
drivers = drivers[['DriverID', 'LastName', 'FirstName', 'ConstructorID', 'Nationality']]
drivers
```
